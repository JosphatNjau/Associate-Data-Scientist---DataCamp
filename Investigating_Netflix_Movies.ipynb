```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load CSV file & inspect the Netflix data
netflix_df = pd.read_csv("netflix_data.csv")
netflix_df.head()
```

```python
# filter the data for movies
movie_df = netflix_df[netflix_df["type"] == "Movie"]

# filter the data for movies released in 1990s
movies_1990s = movie_df[(movie_df["release_year"] >= 1990) & (movie_df["release_year"] < 2000)]
movies_1990s.head()
```

```python
# the most frequent movie duration
plt.hist(movies_1990s["duration"])
plt.title('Disribution of Movie Durations in 1990s')
plt.xlabel('Duration [minutes]')
plt.ylabel('Number of Movies')
plt.show()
```

```python
duration = 100
```

```python
# subset by action movies released in 1990s
action_movies_1990s = movies_1990s[movies_1990s["genre"] == "Action"]

# start the counter
short_movie_count = 0

# iterating through to check if duration < 90 and add 1 to counter
for label, row in action_movies_1990s.iterrows():
    if row["duration"] < 90:
        short_movie_count = short_movie_count + 1
    else:
        short_movie_count = short_movie_count
        
print(short_movie_count)
```
